```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Define the distributions and their parameters
distributions = {
    'Uniform': (stats.uniform, {'loc': -1, 'scale': 2}, 'U(-1, 1)', (-1.5, 1.5)),
    'Uniform+': (stats.uniform, {'loc': 0, 'scale': 1}, 'U(0, 1)', (-0.5, 1.5)),
    'Normal': (stats.norm, {'loc': 0, 'scale': 1}, 'N(0, 1)', (-4, 4)),
    'Exponential': (stats.expon, {'scale': 1/0.2}, 'Exp(0.2)', (0, 15)),
    'Cauchy': (stats.cauchy, {}, 'Cauchy(0, 1)', (-10, 10)),
    'Log-normal': (stats.lognorm, {'s': 1}, 'LogNormal(0, 1)', (0, 6)),
}

# Create a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Analytical Probability Distributions', fontsize=16)

for (name, (dist, params, label, xlim)), ax in zip(distributions.items(), axs.flatten()):
    x = np.linspace(*xlim, 1000)
    y = dist.pdf(x, **params)
    
    ax.plot(x, y, 'b-', lw=2)
    ax.set_title(f'{name} Distribution: {label}')
    ax.set_xlabel('x')
    ax.set_ylabel('Probability Density')
    ax.grid(True, linestyle='--', alpha=0.7)
    
    # Set y-axis to scientific notation if max y > 1000
    if max(y) > 1000:
        ax.ticklabel_format(axis="y", style="sci", scilimits=(0,0))

# Add a "constant" distribution (formerly "No weights")
ax = axs.flatten()[-1]
x = np.linspace(0, 1, 1000)
y = np.ones_like(x)
ax.plot(x, y, 'b-', lw=2)
ax.set_title('Constant Distribution: f(x) = 1')
ax.set_xlabel('x')
ax.set_ylabel('Probability Density')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1.5)
ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig("analytical_distributions_journal.png", dpi=300, bbox_inches='tight')
plt.show()
```
